```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('housing.csv', encoding='latin-1')
```

```python
data
```

```python
data.info()
```

```python
data.dropna(inplace = True)
```

```python
from sklearn.model_selection import train_test_split

x = data.drop(['price'], axis=1)
y = data['price']
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
train_data = x_train.join(y_train)
```

```python
train_data
```

```python
train_data.hist(figsize=(15, 8))
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot=True, cmap='YlGnBu')
```

```python
train_data['rooms'] = np.log(train_data['rooms'] + 1)
train_data['sq'] = np.log(train_data['sq'] + 1)

```

```python
train_data.hist(figsize=(15,8))
```

```python
train_data.city.value_counts()
```

```python
train_data = train_data.join(pd.get_dummies(train_data.city)).drop(['city'], axis=1)
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot=True, cmap='YlGnBu')
```
